```python
'''
all statistic
'''
All_classification = ['all']
bandshort = ['de','th','al','be','ga','hi','all']
band = ['delta','theta','alpha', 'beta','gamma','highgamma','all']
net='COH'

for domain in All_classification:
    for i in range(10):
        globals()[f'Graph_{bd}_train_{i}_'] = []
        # globals()[f'NBS_{bd}_train_{i}'] = []
        df= pd.concat([globals()[f'{domain}_{net}_{bd}_X_train_{i}'], globals()[f'{domain}_{net}_{bd}_y_train_{i}']], axis=1)
        graph_theoretical = df.columns[~df.columns.str.contains('-')]
        # network_weights = df[df.columns[df.columns.str.contains('-')]]
        
        for ft in graph_theoretical:
            if ft=='label':
                pass
            else:
                pval = pg.ttest(df.loc[df['label'] == 1][ft], df.loc[df['label'] == -1][ft])['p-val'].values[0]
                globals()[f'Graph_{bd}_train_{i}_'].append([ft , pval])
        arr = np.array([s[1] for s in globals()[f'Graph_{bd}_train_{i}_']])

        '''
        select top 5 percent
        '''
        k = int(len(arr) * 0.05)
        indices_of_smallest_5_percent = list(np.argsort(arr)[:k])
        print(indices_of_smallest_5_percent)
        globals()[f'Graph_{bd}_train_{i}'] = [a[0] for a in np.array(globals()[f'Graph_{bd}_train_{i}_'])[indices_of_smallest_5_percent]]
        
        '''
        select P<0.05 percent
        '''
        globals()[f'Graph_{bd}_train_{i}_005'] = [x[0] for x in np.array(globals()[f'Graph_{bd}_train_{i}_'])[np.where(arr < 0.05)]]
```

```python

All_classification = ['all']
net='COH'
domain = 'all'
band = ['delta','theta','alpha', 'beta','gamma','highgamma']
bandshort = ['de','th','al','be','ga','hi']

for q in range(10):
    globals()[f'NBS_all_train_{q}'] = []
    df= pd.concat([globals()[f'{domain}_{net}_all_X_train_{q}'], globals()[f'{domain}_{net}_all_y_train_{q}']], axis=1)
    network_weights = df[df.columns[df.columns.str.contains('-')]]
    
    for d, bd in enumerate(band):
        
        pos_index = [i for i in network_weights.index if i < 708]
        neg_index = np.array([i for i in network_weights.index if i >= 708])-(708+639)
        
        globals()[f'positive_{bd}_train_{q}'] = globals()[f'positive_{bd}_EEG'][pos_index]
        globals()[f'negative_{bd}_train_{q}'] = globals()[f'negative_{bd}_EEG'][neg_index]
        
        _, adj, _ = bc.nbs.nbs_bct(globals()[f'negative_{bd}_train_{q}'].T, globals()[f'positive_{bd}_train_{q}'].T, thresh=2.6, tail='both')
        
        matri = np.triu(adj)
        a,b = np.nonzero(matri)
        
        for s in range(len(a)):
            globals()[f'NBS_all_train_{q}'].append('{}_{}-{}_coh'.format(bandshort[d], ch_names[a[s]], ch_names[b[s]]))
    
    with open(f'../VREED_DF_DATA/COH/all_feature_select/all/NBS_all_train_{q}.pkl', 'wb') as f:
        pickle.dump(globals()[f'NBS_all_train_{q}'], f, pickle.HIGHEST_PROTOCOL)
    
```
