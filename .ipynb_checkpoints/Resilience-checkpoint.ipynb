```python
'''
통계분석

'''
xset =np.round((np.arange(1,60)/59)[::-1],2) * 100

import pingouin as pg
from statsmodels.stats.multitest import fdrcorrection

def statistical_resilience(df, resilience_type, method):
    '''
    resilience_type ==> G_E or L_C
    method ==> target or random
    
    '''
    # data upload
    band = ['delta','theta','alpha', 'beta','gamma','highgamma']
    globals()['{}_re_anova_{}'.format(resilience_type, method)] = {'delta':{},'theta':{},'alpha':{},'beta':{},'gamma':{},'highgamma':{}} 
    globals()['{}_re_Ttest_{}'.format(resilience_type, method)] = {'delta':{},'theta':{},'alpha':{},'beta':{},'gamma':{},'highgamma':{}}
    df = globals()['{}_re_{}'.format(resilience_type, method)]
    
    # statistic 
    for b in band:
        for ch in xset:
            #ANOVA 적용
            a = pg.welch_anova(data = df.loc[(df['band']==b) & (df['percent']==ch)], dv = 'value' ,between ='valence')
            #print(a)
            
            if a['p-unc'][0] < 0.05:
                pw = pg.pairwise_gameshowell(data = df.loc[(df['band']==b) & (df['percent']==ch)], dv = 'value' ,between ='valence')
                pw['p-unc'], pw['p-corr'] = fdrcorrection(pw['pval'])
                pwtest = {'A': pw['p-corr'][0],'B': pw['p-corr'][1],'C': pw['p-corr'][2]}
                #print(pw)
                if (pwtest['A']>0.05) and (pwtest['B']>0.05) and (pwtest['C']>0.05):
                    #print('{}_re_anova_{}_{}_{}'.format(resilience_type, method,b,ch))
                   # globals()['{}_re_anova_{}'.format(resilience_type, method)][b][ch] = pwtest
                    pass
                else:
                    #print('{}_re_anova_{}_{}_{}'.format(resilience_type, method,b,ch))
                    globals()['{}_re_anova_{}'.format(resilience_type, method)][b][ch] = pwtest
                    #print(globals()['{}_re_anova_{}'.format(resilience_type, method)][b][ch])
                    
                    ccT = {}

                    if pwtest['A'] < 0.05:
                        ccT['A'] = pwtest['A']
                    else:
                        ccT['A'] = 0

                    if pwtest['B'] < 0.05:
                        ccT['B'] = pwtest['B']
                    else:
                        ccT['B'] = 0

                    if pwtest['C'] < 0.05:
                        ccT['C'] = pwtest['C']
                    else:
                        ccT['C'] = 0

                    globals()['{}_re_Ttest_{}'.format(resilience_type, method)][b][ch] = ccT

            elif a['p-unc'][0] > 0.05:
                ttest = {'A':0,'B':0,'C':0}
                globals()['{}_re_Ttest_{}'.format(resilience_type, method)][b][ch] = ttest
                
    return globals()['{}_re_anova_{}'.format(resilience_type, method)], globals()['{}_re_Ttest_{}'.format(resilience_type, method)] 
```

```python

def ststistic(df, resilience_type, method):
 

    band = ['delta','theta','alpha', 'beta','gamma','highgamma']
    globals()['{}_re_anova_{}_auc'.format(resilience_type, method)] = {'delta':{},'theta':{},'alpha':{},'beta':{},'gamma':{},'highgamma':{}} 
    globals()['{}_re_Ttest_{}_auc'.format(resilience_type, method)] = {'delta':{},'theta':{},'alpha':{},'beta':{},'gamma':{},'highgamma':{}}
    # df = globals()['{}_re_{}_auc'.format(resilience_type, method)]

    a = pg.welch_anova(data = df.loc[(df['band']==b)], dv = 0, between ='valence')

    if a['p-unc'][0] < 0.05:
        pw = pg.pairwise_gameshowell(data = df.loc[(df['band']==b)], dv = 0 ,between ='valence')
        pw['p-unc'], pw['p-corr'] = fdrcorrection(pw['pval'])
        pwtest = {'A': pw['p-corr'][0],'B': pw['p-corr'][1],'C': pw['p-corr'][2]}

        if (pwtest['A']>0.05) and (pwtest['B']>0.05) and (pwtest['C']>0.05):
                pass

        else:
            globals()['{}_re_anova_{}_auc'.format(resilience_type, method)][b] = pwtest
            ccT = {}

            if pwtest['A'] < 0.05:
                ccT['A'] = pwtest['A']
            else:
                ccT['A'] = 0

            if pwtest['B'] < 0.05:
                ccT['B'] = pwtest['B']
            else:
                ccT['B'] = 0

            if pwtest['C'] < 0.05:
                ccT['C'] = pwtest['C']
            else:
                ccT['C'] = 0

            globals()['{}_re_Ttest_{}_auc'.format(resilience_type, method)][b]= ccT

    elif a['p-unc'][0] > 0.05:
        ttest = {'A':0,'B':0,'C':0}
        globals()['{}_re_Ttest_{}_auc'.format(resilience_type, method)][b] = ttest

    return globals()['{}_re_anova_{}_auc'.format(resilience_type, method)], globals()['{}_re_Ttest_{}_auc'.format(resilience_type, method)] 

            

```

```python
ge_re_anova_target_auc, ge_re_Ttest_target_auc = ststistic(G_E_re_auc, 'G_E', 'target' )
ge_re_anova_random_auc, ge_re_Ttest_random_auc = ststistic(G_E_re_random_auc, 'G_E', 'random' )
lc_re_anova_target_auc, lc_re_Ttest_target_auc = ststistic(L_C_re_auc, 'L_C', 'target' )
lc_re_anova_random_auc, lc_re_Ttest_random_auc = ststistic(L_C_re_random_auc, 'L_C', 'random' )
```
